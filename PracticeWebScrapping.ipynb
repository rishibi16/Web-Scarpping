```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.goodreads.com/list/show/1.Best_Books_Ever"
result = requests.get(url)
result.text
```

```python
soup = BeautifulSoup(result.text, 'html5lib')
print(soup.prettify())
```

```python
soup.find("tr").img.get('alt')
```

```python
title = [x.img.get('alt') for x in soup.findAll('tr')]
```

```python
import pandas as pd
data = pd.DataFrame(title, columns=['Title'])
```

```python
data.head()
```

```python
data.loc[1]
```

```python
soup.find('div', class_ = "authorName__container").a.get('href').split('.')[-1]
```

```python
author_name = [x.a.get('href').split('.')[-1] for x in soup.findAll('div', class_ = 'authorName__container')]
author_name
```

```python
author_name = pd.DataFrame(author_name, columns=['Author_name'])
author_name
```

```python
soup.find('span', class_="minirating").text
```

```python
rating = [x.text for x in soup.findAll('span', class_='minirating')]
rating = pd.DataFrame(rating, columns=['Rating'])
```

```python
rating.head()
```
